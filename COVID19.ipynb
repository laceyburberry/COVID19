```python
countries = list(cglobal_aggregated.index)
max_infection_rates = []  
for c in countries :     
    max_infection_rates.append(cglobal_aggregated.loc[c].diff().max())

max_infection_rates
```

```python
cglobal_aggregated["max_infection_rate"] = max_infection_rates
```
